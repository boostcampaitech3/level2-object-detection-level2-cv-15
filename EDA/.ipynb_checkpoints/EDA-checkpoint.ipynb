```python
import pandas as pd
import numpy as np
import json
from pycocotools.coco import COCO
from collections import Counter
import matplotlib.pyplot as plt
import wandb
import random
import os
```

# Json File Load

```python
with open(train_json,"r") as train:
    json_file = json.load(train)
```

```python
annotations = json_file["annotations"]
images = json_file["images"]
categories = json_file["categories"]
```

### Class별 Object 갯수

```python
names = categories["name"]
id_list = []
for i in annotations:
    id_list.append(names[i["category_id"]])
ids = Counter(id_list)

plt.figure(figsize=(17,8))
plt.bar(list(ids.keys()), ids.values(), color = "lightcoral",width=0.7, edgecolor="black",linewidth=2)
for i, value in zip(ids.keys(), ids.values()):
    plt.text(i,value+50, value, ha="center", fontweight = "bold", size = 13)
plt.rc("font",size=10)
plt.rc("xtick", labelsize=16)
# plt.show()
```

```python
train_coco = COCO(train_json)
```

### BBOX

```python
center = {}
bboxs = {}
for i in list(categories["name"]):
    center[i] = []
    bboxs[i] = []
    
for i in train_coco.getAnnIds():
    c = categories["name"][(train_coco.loadAnns(i)[0]["category_id"])]
    x_min = train_coco.loadAnns(i)[0]["bbox"][0]
    y_min = train_coco.loadAnns(i)[0]["bbox"][1]
    w = train_coco.loadAnns(i)[0]["bbox"][2]
    h = train_coco.loadAnns(i)[0]["bbox"][3]
    bboxs[c].append((w,h))
    center[c].append((x_min + w/2 , y_min + h/2))
    
    
```

```python
color_list = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9']
fig, axes = plt.subplots(5,2, figsize=(20,50))
axes = axes.flatten()
for n, i in enumerate(list(categories["name"])):
    for k in center[i]:
        axes[n].scatter(k[0],k[1],color = color_list[n], s = 4)
    axes[n].set_title(i)
plt.show()
```

### BBOX W, H 비율

```python
fig, axes = plt.subplots(5,2, figsize=(20,50))
axes = axes.flatten()
for n, i in enumerate(list(categories["name"])):
    for k in bboxs[i]:
        axes[n].scatter(k[0],k[1],color = color_list[n], s = 4)
    axes[n].set_title(i)
plt.show()
```

# BOX outlier 탐색

```python
bboxs_size = {}
for i in bboxs:
    for k in bboxs[i]:
        if bboxs_size.get(i):
            bboxs_size[i].append(int(k[0]*k[1]))
        else:
            bboxs_size[i] = [int(k[0]*k[1])]
```

```python
bboxs_size_avg = {}
for i in bboxs_size:
    bboxs_size_avg[i] = int(sum(bboxs_size[i])/len(bboxs_size[i]))
```

```python
bboxs_size_avg
```

```python
plt.figure(figsize=(10,5))
plt.bar(bboxs_size_avg.keys(), bboxs_size_avg.values(),color = "lightcoral",width=1, edgecolor="black",linewidth=2)
plt.rc("font",size=10)
plt.show()
```

```python
for i in bboxs_size:
    print(f"{i}'s min size : '{min(bboxs_size[i])}")
```

```python
c = 0
for i in bboxs_size:
    criteria = 300
    count = []
    for k in bboxs_size[i]:
        if k <= criteria:
            count.append(k)
            c += 1
    print(i," : ", len(count))
print(f"Tota : {c}")
```

### 기준 크기보다 작은 박스 시각화

```python
# 가로 & 세로 10 미만 
min_area_images = []
min_area_annotations = []
for i in annotations:
    w = i["bbox"][2] 
    h = i["bbox"][3]
    area = w*h
    if area <= 500:
        min_area_images.append(i["image_id"])
        min_area_annotations.append(i["id"])
min_area_images = list(set(min_area_images))
print(min_area_annotations)
print("=="*30)
print(f"Total_images : {len(min_area_images)}")
print((f"Total_annotations : {len(min_area_annotations)}"))
```

```python
criteria = 500

wandb.init(project="visualization", entity="boostcampaitech3")
labels_file =  "../detection/dataset/train.json"
with open(labels_file) as f:
    labels = json.load(f)

categories = {c["id"]:c["name"] for c in labels["categories"]}
# Get min_areas's image
img_metadata = []
for i in min_area_images:
    i = labels["images"][i]
    img_metadata.append((i["id"], i["file_name"]))

data_path = "../detection/dataset/"
imgs = []
for img_id, img_name in img_metadata:
    wandb_boxes = {"predictions": 
        {
            "box_data":[], 
            "class_labels": categories
        }
    }
    annots = [
        a for a in labels["annotations"] if a["image_id"]==img_id
    ]
    for annot in annots:
        tlx, tly, w, h = annot['bbox']
        if w * h <= criteria:
            cat = str(img_id)
            curr_box = {}
            curr_box["position"] = {
                "minX": tlx, "maxX": tlx+w, 
                "minY": tly, "maxY": tly+h}
            curr_box["class_id"] = cat
            curr_box["box_caption"] = categories[cat]
            curr_box["domain"] = "pixel"
            curr_box["scores"] = { "score" : random.random() }
            wandb_boxes["predictions"]["box_data"].append(curr_box)

    img_path = os.path.join(data_path, img_name)
    img = wandb.Image(img_path, boxes=wandb_boxes)
    imgs.append(img)

wandb.log({"Example": imgs})
# Log onto the Weights & Biases website to view your images
```
